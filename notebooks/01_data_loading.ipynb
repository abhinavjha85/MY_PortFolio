```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print("Libraries loaded! ✅")
```

```python
df = pd.read_csv('/Users/abhinavjha/NeuralRetail/data/raw/online_retail.csv')
print(f"Shape: {df.shape} ✅")
df.head()
```

```python
print(df.columns.tolist())
```

```python
df = df.dropna(subset=['CustomerID'])
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
df = df[df['Quantity'] > 0]
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
print(f"Clean data shape: {df.shape} ✅")
df.to_csv('/Users/abhinavjha/NeuralRetail/data/processed/clean_data.csv', index=False)
print("Clean data saved! ✅")
df.head()
```

```python
print("=== Sales Analysis ===")
print(f"Total Revenue: £{df['TotalPrice'].sum():,.2f}")
print(f"Total Orders: {df['InvoiceNo'].nunique():,}")
print(f"Total Customers: {df['CustomerID'].nunique():,}")
print(f"Total Products: {df['StockCode'].nunique():,}")
print(f"Countries: {df['Country'].nunique()}")
```
